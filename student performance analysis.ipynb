```
# lets import the scipy package
import scipy.stats as stats
import math

# lets seed the random values
np.random.seed(10)

# lets take a sample size
sample_size = 1000
sample = np.random.choice(a= std['math score'],
                          size = sample_size)
sample_mean = sample.mean()

# Get the z-critical value*
z_critical = stats.norm.ppf(q = 0.95)  

 # Check the z-critical value  
print("z-critical value: ",z_critical)                                

# Get the population standard deviation
pop_stdev = std['math score'].std()  

# checking the margin of error
margin_of_error = z_critical * (pop_stdev/math.sqrt(sample_size)) 

# defining our confidence interval
confidence_interval = (sample_mean - margin_of_error,
                       sample_mean + margin_of_error)  

# lets print the results
print("Confidence interval:",end=" ")
print(confidence_interval)
print("True mean: {}".format(std['math score'].mean()))
```

```

# creating an encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# label encoding for test preparation course
std['test preparation course'] = le.fit_transform(std['test preparation course'])

# label encoding for lunch
std['lunch'] = le.fit_transform(std['lunch'])

# label encoding for race/ethnicity
# we have to map values to each of the categories
std['race/ethnicity'] = std['race/ethnicity'].replace('group A', 1)
std['race/ethnicity'] = std['race/ethnicity'].replace('group B', 2)
std['race/ethnicity'] = std['race/ethnicity'].replace('group C', 3)
std['race/ethnicity'] = std['race/ethnicity'].replace('group D', 4)
std['race/ethnicity'] = std['race/ethnicity'].replace('group E', 5)
std['race/ethnicity'].value_counts()
```

```
std['parental level of education'] = le.fit_transform(std['parental level of education'])

#label encoding for gender
std['gender'] = le.fit_transform(std['gender'])

# label encoding for pass_math
std['pass_math'] = le.fit_transform(std['pass_math'])

# label encoding for pass_reading
std['pass_reading'] = le.fit_transform(std['pass_reading'])

# label encoding for pass_writing
std['pass_writing'] = le.fit_transform(std['pass_writing'])

# label encoding for status
std['status'] = le.fit_transform(std['status'])
```

```
# splitting the dependent and independent variables

x = std.iloc[:,:14]
y = std.iloc[:,:14]

print(x.shape)
print(y.shape)
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 45)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```
# importing the MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

# creating a scaler
mm = MinMaxScaler()

# feeding the independent variable into the scaler
x_train = mm.fit_transform(x_train)
x_test = mm.transform(x_test)
```

```
from sklearn.ensemble import RandomForestClassifier

# creating a model
model = RandomForestClassifier()

# feeding the training data to the model
model.fit(x_train, y_train)
model.fit(x_test, y_test)

# predicting the x-test results
y_pred = model.predict(x_test)

# calculating the accuracies
print("Accuracy :", model.score(x_train))
print("Accuracy :", model.score(y_train))
print("Testing Accuracy :", model.score(x_test, y_test))
```

```
# printing the confusion matrix

from sklearn.metrics import confusion_matrix

# creating a confusion matrix
cm = confusion_matrix(y_test, y_pred)

# printing the confusion matrix
plt.rcParams['figure.figsize'] = (8, 8)
sns.heatmap(cm, annot = True, cmap = 'Reds')
plt.title('Confusion Matrix for Random Forest', fontweight = 30, fontsize = 20)
plt.show()
```
